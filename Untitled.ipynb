```python
from gensim.models import Word2Vec
%matplotlib inline
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralClustering
```

```python
nombre = input('modelo:')
cls = int(input('clusters:'))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
modelo = Word2Vec.load(nombre)
lista = list(modelo.wv.vocab.keys())
spectral = SpectralClustering(n_clusters = cls, eigen_solver='arpack', affinity="nearest_neighbors")
spectral.fit(modelo.wv[modelo.wv.vocab.keys()])
label = spectral.labels_

for i in range(len(lista)):
    if(label[i]==0):
        c = 'k'
    elif(label[i]==1):
        c = 'c'
    elif(label[i]==2):
        c = 'm'
    elif(label[i]==3):
        c = 'b'
    elif(label[i]==4):
        c = 'y'
    else:
        continue
    ax.scatter(modelo.wv[lista[i]][0],modelo.wv[lista[i]][1],modelo.wv[lista[i]][2], c = c, marker = 'x')
plt.show()
```
